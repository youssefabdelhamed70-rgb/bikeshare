```python
Start_Duration = df['Start_Station_Name'].value_counts().to_frame().head(15)
Start_Duration['mean'] = df['Trip_Duration'].groupby(df['Start_Station_Name']).mean()
```

```python
fig , ax1 = plt.subplots(figsize=(12,7))
ax1.bar(Start_Duration.index ,
              Start_Duration['count'] ,
              color='blue',
              alpha=0.7,
              label='Count')
ax1.set_ylabel('count',
                color='blue' )

ax1.tick_params(axis='y' ,
                labelcolor='blue')
ax1.set_xticklabels(Start_Duration.index,
                    rotation=45,
                    ha='right') 

ax2 = ax1.twinx()
ax2.plot(Start_Duration.index,
         Start_Duration['mean'],
         color='red',
         marker='o',
         linewidth=2,
         label='Mean')
ax2.set_ylabel('mean' ,
               color='red')
ax2.tick_params(axis='y' ,
                labelcolor='red')

plt.show()
```
